Grouping

```python
#iterate over group-by object
for key, val in df.groupby('Edition'):
    print(key)
    print(val)
```

```python
df.groupby(['Edition','NOC','Medal']).size()
```

```python
df.loc[df.Athlete == 'LEWIS, Carl'].groupby('Athlete').agg({'Edition': ['min', 'max', 'count']})
```

```python
df.groupby('Edition').size().plot(kind='bar');
```

Reshaping

```python
track = df[(df.Edition == 2008) & ( (df.Event == '100m') | (df.Event == '200m'))]
```

```python
track2 = track.groupby(['NOC', 'Gender', 'Discipline', 'Event']).size()
```

```python
track2
```

```python
track3 = track2.unstack()
```

```python
track3
```

```python
#.stack() the dafault value .dropna is set to true - visualise it makes the df taller/longer/more rows
track3.stack('Event')
```

```python
#.unstack() makes df wider/more columns
track3.unstack('Gender', fill_value=0)
```

```python
df[(df.NOC == 'USA') & (df.Medal == 'Gold')].groupby(['Edition', 'Gender']).size().unstack('Gender', fill_value=0).plot(kind='bar', figsize=(10, 8));
```

```python
medals = df.groupby(['Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
medals.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)[['Gold', 'Silver', 'Bronze']].head().plot(kind='bar');
```

Data Viz

```python
chal = df[df.NOC == 'USA']
chal = chal.groupby(['Edition', 'Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
```

```python
chal['Total'] = chal['Gold']+chal['Silver']+chal['Bronze']
```

```python
chal.reset_index(inplace=True)
ans = [group.sort_values('Total', ascending=False)[:1] for year, group in chal.groupby('Edition')]
```

```python
final = pd.DataFrame()
for i in ans:
    final = final.append(i)
final
```
